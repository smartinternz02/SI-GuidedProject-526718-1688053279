```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the dataset into a dataframe
df = pd.read_csv('drug200.csv')
df.head()
```

```python
# Performing the data pre-processing on the data

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['BP'] = le.fit_transform(df['BP'])
df['Cholesterol'] = le.fit_transform(df['Cholesterol'])
df
```

```python
# Splitting the training and testing data

from sklearn.model_selection import train_test_split

x = df.iloc[:,0:5]
y = pd.get_dummies(df.iloc[:,5:]).values
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
# Building the ANN model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Initializing the seq model
class_model = Sequential()
# Adding the input layer to the model
class_model.add(Dense(5,activation='relu'))
# Adding the 1st hidden layer to the model
class_model.add(Dense(16,activation='relu'))
# Adding the 2nd hidden layer to the model
class_model.add(Dense(16,activation='relu'))
# Adding the 3rd hidden layer to the model
class_model.add(Dense(16,activation='relu'))
# Adding the output layer to the model
class_model.add(Dense(5,activation='softmax'))
```

```python
# Compiling the skeletion with loss and categorical crossentropy

class_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Training the model

class_model.fit(xtrain,ytrain,batch_size=4,epochs=100,validation_data=(xtest,ytest))
```

```python
# Classifying the drug

ypred = class_model.predict([[20,0,2,1,15.88]])
print(ypred)
ypred = np.argmax(ypred)
output = ['Drug Y','Drug A','Drug B','Drug C','Drug X']
output[ypred]
```
